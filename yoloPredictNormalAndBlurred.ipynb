```python
def update_yaml_paths(yaml_text, new_path):
    updated_lines = []
    for line in yaml_text.splitlines():
        if line.strip().startswith(("train:", "val:", "test:")):
            key = line.split(":")[0]
            updated_line = f"{key}: {new_path}"
            updated_lines.append(updated_line)
        else:
            updated_lines.append(line)
    return "\n".join(updated_lines)
```

```python
blurDefocus_Blurred = update_yaml_paths(yaml_content,"newBlurImages/Defocus Blurred")
with open("blurDefocus_Blurred.yaml", "w") as f:
    f.write(blurDefocus_Blurred)

metrics_blurred = model.val(
    data='blurDefocus_Blurred.yaml',
    imgsz=640,
    batch=16,
    cache=False,
    save_json=False
)
```
